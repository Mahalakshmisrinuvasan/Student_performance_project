```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
df = pd.read_csv("C:\\Users\\Mahalakshmi S\\Downloads\\archive\\StudentsPerformance.csv")
df.head()
```

```python
print("Shape of dataset:", df.shape)
df.info()
df.describe()
```

```python
print("Missing values:\n", df.isnull().sum())
print("Duplicate rows:", df.duplicated().sum())
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
df.head()
```

```python
sns.boxplot(x='gender', y='math_score', data=df)
plt.title("Math Score by Gender")
plt.show()
```

```python
sns.violinplot(x='lunch', y='reading_score', data=df)
plt.title("Reading Score by Lunch Type")
plt.show()
```

```python
sns.barplot(x='test_preparation_course', y='writing_score', data=df)
plt.title("Writing Score by Test Prep Course")
plt.show()
```

```python
sns.boxplot(x='parental_level_of_education', y='math_score', data=df)
plt.title("Math Score by Parental Education")
plt.xticks(rotation=45)
plt.show()
```

```python
num_cols = ['math_score', 'reading_score', 'writing_score']
sns.heatmap(df[num_cols].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Between Scores")
plt.show()
```

```python

df['average_score'] = df[num_cols].mean(axis=1)

sns.histplot(df['average_score'], kde=True, bins=20)
plt.title("Distribution of Average Scores")
plt.show()
```

```python
print(" INSIGHTS:")
print("- Students who completed test preparation scored higher.")
print("- Females scored higher in reading and writing.")
print("- Parental education influences student performance.")
print("- Lunch type (standard) leads to better scores.")
```
